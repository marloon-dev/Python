# Listas em Python : 

## Criando uma lista:

    Para criarmos uma lista usaremos o simbolo [] (colchetes). 
    Essa lista precisa ser armazenada em uma VARIÁVEL
    Os itens de uma lista são separados pela virgula

```python
#lista_compras é a variável
#A nossa lista de compras é composta por 3 itens: banana, laranja e maçã
lista_compras = ['banana','laranja','maçã']
print(lista_compras)
```

## Acessando os dados de uma lista:

    Cada item da lista possui uma posição. Por padrão, as posições não iniciam do 1(um) e sim do 0(ZERO)

```python
#Para acessarmos um item da lista vamos utilizar a estrutura: nomedalista[posição]
# Buscando a posição [1]
print(lista_compras[1])
```

```python
# Buscando a posição [0]
print(lista_compras[0])
```

```python
# Se tentarmos a posição [3] teremos um erro. Isso corre pois não existe uma posição 3. Apenas as posições [0], [1], [2]
print(lista_compras[3])
```

```python
#O uso de números negativos permite acessar a lista na ordem inversa
print(lista_compras[-1])
```

## Adicionando um item a lista:

    Para adicionarmos um novo item a lista, usaremos métodos. Um método funciona como uma espécie de função.
    2 métodos possíveis são:
    -.insert()
    -.append()
        
    No exemplo abaixo tentamos incluir o item 'carro' a nossa lista_compras.
    
## Deletando um item da lista:

    Para excluirmos um item da lista podemos usar algumas opções:
    - del
    - .pop()
    - .remove() 

```python
# .append() -> Adiciona AO FIM da lista_compras o item 'carro'
lista_compras.append('carro')
print(lista_compras)
```

```python
# del -> Remove um item da lista baseado na posição indicada. 
del lista_compras[3]
print(lista_compras)
```

```python
# .insert -> Insere um item na lista na posição indicada
lista_compras.insert(1,'carro')
print(lista_compras)
```

```python
lista_compras.append('carro')
```

```python
# .remove() -> Remove um item baseado no seu valor e não na sua posição. 
#ATENÇÃO! retira apenas a primeira ocorrência e não todas!
lista_compras.remove('carro')
```

```python
print(lista_compras)
```

## Criando uma lista do zero:

    É muito comum que as listas se iniciem sem nenhum valor. Como se fosse um papel em branco que aos poucos você vai adicionando informações. 
    Para criarmos vamos usar novamente o colchetes, mas sem nenhum item.

```python
#Lista VAZIA
lista_sonhos = []
#.pop() -> Remove da lista_compras o último item, mas não o exclui. Nesse caso, o valor carro será armazenado na variável item
item = lista_compras.pop(-1)
print(item)
#Após retirar da lista_compras, adicionamos 'carro' na nossa lista_sonhos
lista_sonhos.append(item)
```

```python
print(lista_sonhos)
```

## Criando uma Lista de tarefas (TO DO):

```python
# Criando a lista tarefas 
tarefas = []
# Usando o Input() vamos coletar do usuário qual a tarefa a ser adicionada.
atividade = input('Insira uma atividade')
#Adiciona a tarefa indicada pelo usuário a lista de tarefas
tarefas.append(atividade)
```

```python
print(tarefas)
```

## Inserindo vários itens na lista

```python
while atividade:
    atividade = input('Insira uma atividade')
    tarefas.append(atividade)
```

```python
#A estrutura abaixo indica que pegará os itens da posição [0] até o penúltimo item da lista [-2]
print(tarefas[:-2])
```

```python
#Vamos escrever todas os itens da lista em outro formato. 

#O FOR nos pertmite percorrer todos itens da lista tarefas. 
for tarefa in tarefas:
    print(tarefa)
```

## Criando uma lista com valores numéricos:

    Até agora todas nossas listas foram compostas por textos, mas é possível ter listas com valores númericos. 
    
    No exemplo abaixo temos 3 lojas e seu volume em vendas. 

```python
# Usando números
lojas = ['Rio de Janeiro','São Paulo', 'Curitiba']
vendas = [10000, 20000, 50000] 
```

```python
#.sort() -> Ordena os valores do maior para o menor. 
# ATENÇÃO - > As duas listas não estão conectadas.
lojas.sort()
vendas.sort()
```

```python
#Perceba que ao mudarmos a ordem, não temos mais a Rio de Janeiro -10000 / São Paulo - 20000 / Curitiba - 50000
print(lojas)
print(vendas)
```

## Vamos unir os itens em uma Tupla:

### O que é uma Tupla?

Tupla é um objeto assim como as listas. Ao invés de usarmos o [] usaremos () para criar uma dupla

```python
lojas = ['Rio de Janeiro','São Paulo', 'Curitiba']
vendas = [10000,20000,50000]
```

```python
resultados=[]
# Usando range para rodar o for. i é uma variável auxiliar que irá percorrer a posição 0, 1 e 2 da nossas listas.
# Range(3) indica que são 3 iterações 
for i in range(3):
    tupla=(lojas[i],vendas[i])
    resultados.append(tupla)
```

```python
print(resultados)
```

## Acessando um item dentro da tupla:

    Para acessarmos um item dentro da tupla iremos 1º indicar qual a posição do item da lista e depois a posição da tupla. 
    
    No exemplo abaixo ao fazermos resultados[1] estamos acessando o segundo item da lista -> ('São Paulo', 20000)
    
    Ao fazermos resultados[1][0] estamos acessando o primeiro item da tupla, ou seja: São Paulo

```python
print(resultados[1][0])
```
